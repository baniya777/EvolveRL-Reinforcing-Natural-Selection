```python
import time
import random
import matplotlib.pyplot as plt

a = env = Hex_Env(render_mode="human")
a.reset()
count = 0
terminated = False

x = []
y = []
z = []
# for i in range(count):
while count<15000:
    count +=1
    actions = []
    for agent in a.agents:
        if agent[3] == 'y':
            actions.append((agent,random.randint(1,7)))
        else:
            act = env.checkForPrey(agent)
            if act == 0:
                actions.append((agent,random.randint(1,9)))
            else:
                actions.append((agent,act))
    actions = dict(actions)



    x.append(len(env.possible_predator)+env.new_predators-env.dead_predators)
    y.append(len(env.possible_prey)+env.new_prey-env.dead_preys)
    z.append(env.timestep)
     
    a.step(actions)
    print(env.timestep)
    if env.terminationTrue == True:
        break

    terminated = a.render()
    

    # for event in pygame.event.get():
    #         if event.type == pygame.QUIT:
    #             terminated = True

plt.figure(figsize=(10,6))
plt.plot(z,x,label="Predator (x)")
plt.plot(z,y,label="prey(y)")
plt.xlabel("time")
plt.ylabel("population")
plt.legend()
plt.grid(True)
plt.show()
```
